```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
class svm_classifier():
    def __init__(self, lr, noi, lambdap):
        self.lr = lr
        self.noi = noi
        self.lambdap = lambdap
    
    def fit(self, x, y):
        self.m, self.n = x.shape
        self.w = np.zeros(self.n)
        self.b = 0
        self.x = x
        self.y = y
        for i in range(self.noi):
            self.update_weights()
    def update_weights(self):
        y_label = np.where(self.y <= 0, -1, 1)
        for index, x_i in enumerate(self.x):
            condition = y_label[index]*(np.dot(x_i, self.w) - self.b) >= 1
            if(condition == True):
                dw = 2*self.lambdap*self.w
                db = 0
            else: 
                dw = 2+self.lambdap*self.w - np.dot(x_i, y_label[index])
                db = y_label[index]
            self.w = self.w - self.lr*dw
            self.b = self.b - self.lr*db
    def predict(self, x):
        output = np.dot(x, self.w) - self.b
        return np.where(output >= 0, 1, 0)

```

```python
data = pd.read_csv('gene_expression.csv')
data = data.dropna()
features = data.iloc[:, :-1].values
outcomes = data.iloc[:, -1].values
x_train, x_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.2, random_state=1)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
model = svm_classifier(lr = 0.001, noi = 1000, lambdap = 0.01)
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
acc = accuracy_score(y_test, pred)
print(f'the accuracy score is {acc}')
```
