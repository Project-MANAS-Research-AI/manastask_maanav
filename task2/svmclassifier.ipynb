```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
class svm_classifier():
    def __init__(self, lr, noi, lambdap):
        self.lr = lr
        self.noi = noi
        self.lambdap = lambdap
    
    def fit(self, x, y):
        self.m, self.n = x.shape
        self.w = np.zeros(self.n)
        self.b = 0
        self.x = x
        self.y = y
        for i in range(self.noi):
            self.update_weights()
    def update_weights(self):
        y_label = np.where(self.y <= 0, -1, 1)
        for index, x_i in enumerate(self.x):
            condition = y_label[index]*(np.dot(x_i, self.w) - self.b) >= 1
            if(condition == True):
                dw = 2*self.lambdap*self.w
                db = 0
            else: 
                dw = 2+self.lambdap*self.w - np.dot(x_i, y_label[index])
                db = y_label[index]
            self.w = self.w - self.lr*dw
            self.b = self.b - self.lr*db
    def predict(self, x):
        output = np.dot(x, self.w) - self.b
        return np.where(output >= 0, 1, 0)

```

```python
data = pd.read_csv('archive\\pulsar_data_train.csv')
```

```python
features = data.iloc[:, :-1].values.astype(float)
outcomes = data.iloc[:, -1].values.astype(int)
scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
model = svm_classifier(lr = 0.001, noi = 1000, lambdap = 0.01)
model.fit(features, outcomes)
```

```python
data2 = pd.read_csv('archive\\pulsar_data_train.csv')
features2 = data2.iloc[:, :-1].values.astype(float)
outcomes2 = data2.iloc[:, -1].values.astype(int)
features2 = scaler.fit_transform(features2)

predict2 = model.predict(features2)
acc = accuracy_score(outcomes2, predict2)
print(f'the accuracy is {acc}')
```
